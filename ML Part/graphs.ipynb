```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("comparison.csv",index_col="index")
```

```python
data.head()
```

```python
data["1"] = data["1 up"] + data["1 down"]
data["2"] = data["2 up"] + data["2 down"]
data["3"] = data["3 up"] + data["3 down"]
```

```python
data[["1","2","3"]].rolling(window=100,min_periods=1).mean().plot()
```

```python
data[["1 up","2 up","3 up"]].rolling(window=100,min_periods=1).mean().plot()
```

```python
data[["1 down","2 down","3 down"]].rolling(window=100,min_periods=1).mean().plot()
```
