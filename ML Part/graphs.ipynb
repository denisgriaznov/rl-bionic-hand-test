```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("comparison.csv",index_col="index")
```

```python
data["index"]
```

```python
data.head()
```

```python
data["R, λ=1"] = data["1 up"] + data["1 down"]
data["R, λ=2"] = data["2 up"] + data["2 down"]
data["R, λ=3"] = data["3 up"] + data["3 down"]
```

```python
data["Ro, λ=1"] = data["1 up"] 
data["Ro, λ=2"] = data["2 up"]
data["Ro, λ=3"] = data["3 up"] 
```

```python
data["Rt, λ=1"] = data["1 down"]
data["Rt, λ=2"] = data["2 down"]
data["Rt, λ=3"] = data["3 down"]
```

```python
data[["R, λ=1","R, λ=2","R, λ=3"]].rolling(window=100,min_periods=1).mean().plot()
```

```python
data[["Ro, λ=1","Ro, λ=2","Ro, λ=3"]].rolling(window=100,min_periods=1).mean().plot()
```

```python
data[["Rt, λ=1","Rt, λ=2","Rt, λ=3"]].rolling(window=100,min_periods=1).mean().plot()
```
